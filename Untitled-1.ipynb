```python
df = pd.read_csv("trac2_CONVT_train.csv")
list_convos = []
list_instance = []

convoColumnIndex = df.columns.get_loc("text")
turnColumnIndex = df.columns.get_loc("turn_id")

for i in range(0, len(df) - 1):
    
    if df.iloc[i + 1, turnColumnIndex] > df.iloc[i, turnColumnIndex]:
        list_instance.append(df.iloc[i, convoColumnIndex])
    else:
        list_instance.append(1)
        list_convos.append(list_instance)
        list_instance = []

list_convos[-1].append(df.iloc[len(df) - 1, convoColumnIndex])
print(list_convos)
```
